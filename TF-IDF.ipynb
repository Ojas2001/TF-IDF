```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
A = open("t_1.txt",'r')
B = open("t_2.txt",'r')
C = open("t_3.txt",'r')
doc_A = A.read()
doc_B = B.read()
doc_C = C.read()
bag_of_words_A = doc_A.split()
bag_of_words_B = doc_B.split()
bag_of_words_C = doc_C.split()
bag_of_words_A = pd.DataFrame(bag_of_words_A)
bag_of_words_B = pd.DataFrame(bag_of_words_B)
bag_of_words_C = pd.DataFrame(bag_of_words_C)
```

```python
bag_words = pd.DataFrame([bag_of_words_A,bag_of_words_B,bag_of_words_C])
bag_words
```

```python
unique_words = set(bag_of_words_A).union(set(bag_of_words_B)).union(set(bag_of_words_C))
```

```python
no_words_A = dict.fromkeys(unique_words,0)
for word in bag_of_words_A:
    no_words_A[word] += 1
```

```python
no_words_B = dict.fromkeys(unique_words,0)
for word in bag_of_words_B:
    no_words_B[word] += 1
```

```python
no_words_C = dict.fromkeys(unique_words,0)
for word in bag_of_words_C:
    no_words_C[word] += 1
```

```python
no_words_df = pd.DataFrame([no_words_A,no_words_B,no_words_C])
no_words_df
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_word=stopwords.words('english')
```

```python
def compute_TF(word_Dict,bag_of_words):
    tf_dict = {}
    bag_of_words_count = len(bag_of_words)
    for word, count in word_Dict.items():
        tf_dict[word] = count/float(bag_of_words_count)
    return tf_dict

tf_doc_A = compute_TF(no_words_A,bag_of_words_A)
tf_doc_B = compute_TF(no_words_B,bag_of_words_B)
tf_doc_C = compute_TF(no_words_C,bag_of_words_C)
```

```python
tf_df = pd.DataFrame([tf_doc_A,tf_doc_B,tf_doc_C])
tf_df
```

```python
import math
def compute_IDF(documents):
    n = len(documents)
    
    idf_dict = dict.fromkeys(documents[0].keys(),0)
    for document in documents:
        for word, val in document.items():
            if val>0:
                idf_dict[word] += 1
    
    for word, val in idf_dict.items():
        idf_dict[word] = math.log(n/float(val))
    
    return idf_dict

idfs = compute_IDF([no_words_A,no_words_B,no_words_C])
```

```python
idfs
```

```python
def compute_TF_IDF(tf_bag_of_words,idfs):
    tf_idf = {}
    for word, val in tf_bag_of_words.items():
        tf_idf[word] = val*idfs[word]
    return tf_idf

tf_idf_A = compute_TF_IDF(tf_doc_A,idfs)
tf_idf_B = compute_TF_IDF(tf_doc_B,idfs)
tf_idf_C = compute_TF_IDF(tf_doc_C,idfs)
```

```python
tf_idf_df = pd.DataFrame([tf_idf_A,tf_idf_B,tf_idf_C])
```
